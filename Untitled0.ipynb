```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("kainatjamil12/housing")

print("Path to dataset files:", path)
```

```python
df = pd.read_csv("/root/.cache/kagglehub/datasets/kainatjamil12/housing/versions/1/Housing.csv")

palette = ["#007f5f","#2b9348","#55a630","#80b918","#aacc00",
           "#bfd200","#d4d700","#dddf00","#eeef20","#ffff3f"]
```

```python
df.describe()
```

```python
categorias = df['mainroad'].value_counts().index.to_list()
valores = df['mainroad'].value_counts().values.tolist()

fig, ax = plt.subplots(nrows=1, ncols=1)

grafico_de_barras = ax.bar(
    x=categorias,
    height=valores,
    width=0.9,
    bottom=0,
    color=['#2b9348','#80b918']
)

plt.suptitle(
    "Source: Housing Price Dataset — Factors Affecting Home",
    fontsize=8
)

plt.title("Number of houses with direct access to a main road")

ax.bar_label(grafico_de_barras)

plt.yticks([])

plt.show()
```

```python
categorias = df['guestroom'].value_counts().index.to_list()
valores = df['guestroom'].value_counts().values.tolist()

# Translate yes/no -> readable text
mapa = {
    "no": "No guest room",
    "yes": "Has guest room"
}
categorias_legiveis = [mapa[c] for c in categorias]

fig, ax = plt.subplots(1, 1)

# Create bar chart
graph_guestroom = ax.bar(
        x=categorias_legiveis,
        height=valores,
        color=['#2b9348','#80b918']
)

# Add values on top of bars
ax.bar_label(graph_guestroom)

# Titles
plt.suptitle(
    "Source: Housing Price Dataset — Factors Affecting Home",
    fontsize=8
)
plt.title("Number of houses with a guest room")

# Legend
ax.legend(
    graph_guestroom,
    categorias_legiveis
)

# Remove Y-axis ticks
plt.yticks([])

plt.show()
```

```python
bins = [
    1750000, 2905000, 4060000, 5215000, 6370000,
    7525000, 8680000, 9835000, 10990000, 12145000, 13300000
]

labels = [
    "1,750,000 - 2,905,000",
    "2,905,000 - 4,060,000",
    "4,060,000 - 5,215,000",
    "5,215,000 - 6,370,000",
    "6,370,000 - 7,525,000",
    "7,525,000 - 8,680,000",
    "8,680,000 - 9,835,000",
    "9,835,000 - 10,990,000",
    "10,990,000 - 12,145,000",
    "12,145,000 - 13,300,000"
]

df['price_range'] = pd.cut(
    df['price'],
    bins=bins,
    labels=labels,
    include_lowest=True
)

# Mean
mean_price = df['price'].mean()

# Data for the plot
ranges = df['price_range'].value_counts().index.to_list()
values = df['price_range'].value_counts().values.tolist()

# Figure and axis
fig, ax = plt.subplots(nrows=1, ncols=1)

# Bar chart
graph_price = ax.bar(
    x=ranges,
    height=values,
    color=[
        "#007f5f","#2b9348","#55a630","#80b918",
        "#aacc00","#bfd200","#d4d700","#dddf00",
        "#eeef20","#ffff3f"
    ]
)

# Titles
plt.suptitle(
    "Source: Housing Price Dataset — Factors Affecting Home",
    fontsize=8
)
plt.title("Distribution of houses by price range")

# Rotation to read labels
plt.xticks(rotation=90)

# Remove Y ticks
plt.yticks([])

# Add labels on top of bars
plt.bar_label(graph_price)

# Show
plt.show()
```

```python
categorias = df['basement'].value_counts().index.to_list()
valores = df['basement'].value_counts().values.tolist()

tradutor = { "no": "No Basement", "yes": "Has Basement" }
categorias_legiveis = [str(c) for c in categorias_legiveis]

fig, ax = plt.subplots(1, 1)

graph = ax.bar(
    x=categorias_legiveis,
    height=valores,
    color=palette[:len(categorias)]
)

ax.bar_label(graph)

plt.suptitle("Source: Housing Price Dataset — Factors Affecting Home", fontsize=8)
plt.title("Distribution of Basement")

ax.legend(graph, categorias_legiveis)
plt.yticks([])

plt.show()
```

```python
categorias = df['furnishingstatus'].value_counts().index.to_list()
valores = df['furnishingstatus'].value_counts().values.tolist()

# Tradução sugestão
tradutor = {
    "furnished": "Furnished",
    "semi-furnished": "Semi-Furnished",
    "unfurnished": "Unfurnished"
}
categorias_legiveis = [tradutor.get(c, c) for c in categorias]

fig, ax = plt.subplots(1, 1)

graph = ax.bar(
    x=categorias_legiveis,
    height=valores,
    color=palette[:len(categorias)]
)

ax.bar_label(graph)

plt.suptitle("Source: Housing Price Dataset — Factors Affecting Home", fontsize=8)
plt.title("Distribution of Furnishing Status")

ax.legend(graph, categorias_legiveis)
plt.yticks([])

plt.show()
```
